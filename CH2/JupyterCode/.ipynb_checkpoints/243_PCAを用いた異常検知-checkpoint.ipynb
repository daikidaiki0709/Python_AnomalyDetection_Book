- PCAではローディングおよび主成分スコアを用いて、元データを再構築する（この際に成分数は削減しておく、そのままだとほぼ同じ値が返ってくるため）

- 元データと再構築元データの誤差（再構成誤差）を算出し、そのL2ノルムを算出（この値が異常度となる）
    - X' = X@L.T@Lとし、||X - X'||を算出する
    
- 分位点法（など）で異常度の閾値を算出

- 記号の説明
    - S = X@L.T
    
    - S : n×component
    - L : feature×　component（変換行列）
    - X : n×feature




```python
import numpy as np
import pandas as pd
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
```

```python
X = np.random.multivariate_normal([0,0],[[1,0.5],[0.5,1]],1000)
X =np.insert(X,120,np.array([10,60]),axis=0)
X = np.insert(X, 600, np.array([30,15]),axis=0)
# X = X.reshape(1002,2)
X = X - X.mean(axis=0)
X = pd.DataFrame(X)
```

```python
plt.scatter(X.iloc[:,0],X.iloc[:,1])
```

```python
pca = PCA(n_components=1)
results = pca.fit(X)

# 寄与率
print(pca.explained_variance_ratio_)
print()

# ローディング
print(pca.components_)
```

```python
# 主成分スコアから元データを再構築
X_inverse = pd.DataFrame(pca.inverse_transform(pca.transform(X)))

# 元データとの再構成誤差を算出
X_error = X_inverse - X

# 誤差から異常度を算出
X_alpha = np.sqrt((X_error**2).sum(axis=1))

# 異常度をプロット
X_axis = [i for i in range(X_alpha.shape[0])]

plt.figure(figsize=(18,6))
plt.scatter(X_axis, X_alpha,s=15, c='k')
plt.plot([0,X_axis[-1]], [10,10],linestyle='dashed',linewidth=1,c='r',label='threshold')
plt.xlim(0,1000)
plt.legend()
```
