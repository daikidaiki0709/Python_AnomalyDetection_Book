# 異常検知の流れ


### 1. 異常度の定義
- 異常度（α)は以下で定義される ： α =  -ln(p(x'| D, θ))
### 2. 閾値の設定
- 統一した定量的基準はない（多くは経験則）
- しかし、定量的に基準を設ける手法も存在する
    - （後述）分位点法
    - （後述）ホテリング法
### 3. モデルの検証
- 検証データで精度評価

### 分位点法

- 原理：サンプルデータの総数に対して数％の異常率を設ける

    - 分位点 (%) = (異常データ数）/(全データ数）
    - 異常データ数 = 全データ数×分位点
    
- 実際は異常度の大きさでデータをソートし、異常データ数分のデータを除去する

```python
# coding: utf-8

import matplotlib.pyplot as plt
import warnings
import numpy as np
import pandas as pd
import scipy as sp
import scipy.stats
from numpy import linalg as la
from scipy.stats import norm
import seaborn as sns

warnings.simplefilter('ignore')

### データ読み込み ###
davis = pd.read_csv('./../data/Davis.csv').values
x = davis[: ,2:3]


### データの概要を可視化 ###
sns.distplot(x, fit = norm, color='k', kde=False, bins =50,rug=True)
plt.xlabel('体重[kg]')
plt.ylabel('確率密度')
plt.show()


###  異常度の算出と可視化 ##

# データの中心化（平均を0とする）
xc = x - np.mean(x)

# データの分散
sx = (1/len(x))*(xc.T@xc)

# 異常度の定義（正規分布を仮定とする）
ap = xc**2 / sx

### 分位点法を用いた閾値設定 ###
THRETH_PERCENT = 0.05
THRETH = int(len(x)*THRETH_PERCENT)

# 異常度の分布を可視化
fig = plt.figure(figsize=(16,8))
plt.hist(ap,color='blue',bins=200)
plt.ylabel('異常度',fontsize=30)
plt.plot([ap_detect,ap_detect], [0,100], color='red', linestyle='-', linewidth=1, label='閾値')
plt.xticks(fontsize=24)
plt.yticks(fontsize=24)
plt.legend(fontsize=24)
plt.show()
fig.savefig('./../data/異常度.png',dpi=360)

# 異常度配列をsortする
ap_sort = np.sort(ap,axis=0)
ap_detect = ap_sort[-1*THRETH:][0]

plt.scatter(np.arange(ap.size), ap , color='b')
plt.plot([0,200], [ap_detect,ap_detect] , color='red', linestyle='-', linewidth=1, label='threshold')
plt.ylim(0,50)
plt.show()
```

### ホテリング法
- 分位点に基づき、カイ二乗分布から閾値を設定する
    - 前提：データが正規分布に従っている
    - 異常度はデータ数が十分に大きい場合、自由度1のカイ二乗分布に従う

```python
# coding: utf-8

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import scipy
import scipy.stats
from numpy import linalg

### データの概要を可視化 ###
sns.distplot(x, fit = norm, color='k', kde=False, bins =50,rug=True)
plt.xlabel('体重[kg]')
plt.ylabel('確率密度')
plt.show()


###  異常度の算出と可視化 ##
# データの中心化（平均を0とする）
xc = x - np.mean(x)

# データの分散
sx = (1/len(x))*(xc.T@xc)

# 異常度の定義（正規分布を仮定とする）
ap = xc**2 / sx

# 閾値
gamma = 0.02
th = scipy.stats.chi2.ppf(1-gamma,1) # カイ二乗分布の累積確率が2%となる横軸の値

plt.scatter(np.arange(ap.size), ap , color='b')
plt.plot([0,200], [th,th] , color='red', linestyle='-', linewidth=1)
plt.ylim(0,55)
plt.show()
```
