### 多次元版のホテリング法

```python
# coding: utf-8

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import scipy as sp
import scipy.stats
from numpy import linalg as la
from scipy.stats import norm
import seaborn as sns

from matplotlib import cm

### データ読み込みと可視化 ###
# データの読み込み
davis = pd.read_csv('./../data/Davis.csv').values
d = davis[: ,2:4]

# 元データのプロット
plt.scatter( d[:,0], d[:,1],color='black')
plt.xlim(20,180)
plt.ylim(40,210)
plt.xlabel('体重[kg]')
plt.ylabel('身長[cm]')


# ２変数正規分布を表示する

# 2変数meshgridの作成
x =  np.arange(20, 180,1)
y =  np.arange(40, 210,1)
X, Y = np.meshgrid(x, y)

# 中心化データ
xc = d - d.mean(axis = 0)

# 平均ベクトル
mx = d.mean(axis = 0)

# 分散共分散行列
sx = 1/(d.shape[0])*(xc.T@xc).astype('float')

#  2変数混合正規分布の等高線を描画
f = lambda x, y: scipy.stats.multivariate_normal(mx, sx).pdf([x, y]) # ２変数正規分布の確率密度を示す関数
Z = np.vectorize(f)(X, Y)

plt.contour(X, Y, Z,levels=10,linewidths=1)
plt.show()



###  異常度の算出と可視化 ##

# 異常度
ap = (1/2)*(xc@np.linalg.inv(sx))*xc
'''
apは各変数ごとの異常度を格納している
最終的にはサンプルごとに異常度を１つの値にする必要がある（それか各変数ごとに異常サンプルを抽出するか）
今回は単純な異常度の和で評価する
'''

# 各変数における異常度の総和
a = ap[:,0] + ap[:,1]

# 可視化
gamma = 0.05
th = sp.stats.chi2.ppf(1-gamma,1)
plt.scatter(np.arange(a.size), a , color='b')
plt.plot([0,200], [th,th] , color='red', linestyle='-', linewidth=1)
plt.show()
```
